```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
img=np.asarray(Image.open("ROSE picture.JPG"))
lum_img=img[:, :, 0]
plt.imshow(lum_img).set_cmap("nipy_spectral")
plt.colorbar()
plt.show()
```
